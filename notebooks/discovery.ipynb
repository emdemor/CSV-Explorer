```python
import matplotlib.pyplot as plt
```

```python
fig = plt.gcf()

# Verifica se há eixos na figura
if not fig.axes:
    print("A figura está vazia.")
else:
    print("A figura contém eixos/gráficos.")
```

```python
fig.show()
```

```python
def has_content(fig):
    has_content = False
    for ax in fig.axes:
        if ax.lines or ax.patches or ax.texts or ax.images or ax.collections:
            has_content = True
            break
    return has_content
```

```python
has_content(fig)
```

```python
import pandas as pd
df = pd.read_csv((config.EXAMPLES_PATH / "clients.csv").as_posix())
```

```python
import matplotlib.pyplot as plt

# Count the frequency of each age group
age_counts = df['Faixa_Etária'].value_counts()

# Plot the distribution of age groups
plt.figure(figsize=(10,6))
plt.bar(age_counts.index, age_counts.values, color='skyblue')
plt.xlabel('Faixa Etária')
plt.ylabel('Frequência')
plt.title('Distribuição de Faixa Etária')
plt.xticks(rotation=45)
plt.show()
```
